```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
  def __init__(self, dataframe):
    self.dataframe = dataframe
    self.X = dataframe.iloc[:,:-1].to_numpy()
    self.y = dataframe.iloc[:,-1].to_numpy()
    self.X_tensor = torch.from_numpy(self.X).type(torch.float)
    self.y_tensor = torch.from_numpy(self.y).type(torch.float)

  def __len__(self):
    return len(self.dataframe)

  def __getitem__(self,idx):
    return nn.functional.normalize(self.X_tensor[idx],dim=0), self.y_tensor[idx]

```

```python
dataset = CustomDataset(df)

train_data, test_data = torch.utils.data.random_split(dataset,[0.8,0.2])

train_dataloader = DataLoader(train_data,32,True)
test_dataloader = DataLoader(test_data,32,False)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
num_features = train_data.dataset.X.shape[1]
num_features
```

```python
class ObesityModelV0(nn.Module):
  def __init__(self, in_features, hidden_units, out_features):
    super().__init__()
    self.layers_stack = nn.Sequential(
        nn.Linear(in_features,hidden_units),
        nn.ReLU(),
        nn.Linear(hidden_units,hidden_units),
        nn.ReLU(),
        nn.Linear(hidden_units,hidden_units),
        nn.ReLU(),
        nn.Linear(hidden_units,hidden_units),
        nn.ReLU(),
        nn.Linear(hidden_units,out_features),

    )

  def forward(self , x):
    return self.layers_stack(x)
```

```python
model_0 = ObesityModelV0(num_features,20,len(class_names)).to(device)
model_0
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau
```

```python
accuracy = Accuracy(task='multiclass', num_classes=len(class_names)).to(device)
```

```python
def train_step(model,dataloader,loss_fn,optimizer,scheduler):
  model.train()
  train_loss, train_acc = 0,0
  for batch, (X,y) in enumerate(dataloader):
    X,y = X.to(device), y.type(torch.LongTensor).to(device)
    y_pred = model(X)
    loss = loss_fn(y_pred,y)
    acc = accuracy(y_pred,y)
    train_loss += loss
    train_acc += acc
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  train_loss /= len(dataloader)
  train_acc /= len(dataloader)
  scheduler.step(train_loss)

  return train_loss, train_acc
```

```python
def test_step(model, dataloader,loss_fn):
  model.eval()
  test_loss, test_acc = 0,0
  with torch.inference_mode():
    for batch, (X,y) in enumerate(dataloader):
      X,y = X.to(device), y.type(torch.LongTensor).to(device)
      test_pred = model(X)
      loss = loss_fn(test_pred,y)
      acc = accuracy(test_pred,y)
      test_loss+=loss
      test_acc+=acc
    test_loss /= len(dataloader)
    test_acc /= len(dataloader)
    return test_loss,test_acc
```

```python
from tqdm.auto import tqdm

train_losses,test_losses,train_accs,test_accs = [],[],[],[]

def train(model,train_dataloader,test_dataloader,loss_fn,optimizer,scheduler,epochs):

  for epoch in tqdm(range(epochs)):
    train_loss,train_acc = train_step(model,train_dataloader,loss_fn,optimizer,scheduler)
    test_loss,test_acc = test_step(model,test_dataloader,loss_fn)
    train_losses.append(train_loss.detach().cpu().numpy())
    test_losses.append(test_loss.detach().cpu().numpy())
    train_accs.append(train_acc.detach().cpu().numpy())
    test_accs.append(test_acc.detach().cpu().numpy())

    print(
        f"Epoch:{epoch+1} | "
        f"Train Loss:{train_loss:.3f} | "
        f"Train Acc:{train_acc*100:.3f}% | "
        f"Test Loss:{test_loss:.3f} | "
        f"Test Acc:{test_acc*100:.3f}%"
    )
```

```python
NUM_EPOCHS = 500

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_0.parameters())
scheduler = ReduceLROnPlateau(optimizer)

from timeit import default_timer as timer
start = timer()
train(model_0,train_dataloader,test_dataloader,loss_fn,optimizer,scheduler,NUM_EPOCHS)
end = timer()
print(f"Total training time:{end-start:.3f} seconds")

```
